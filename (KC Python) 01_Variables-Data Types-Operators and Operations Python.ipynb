```python
num_1: int = 7 # Assignment operator
num_2: int = 2

print(f'num1  = {num_1}')

num_1: int = 7
num_2: int = 2
num_1 += num_2 # num_1 = num_1 + num_2 = 2 + 7
print(f'num_1 += num_2 = {num_1}') # Additional assignment

num_1: int = 7
num_2: int = 2
num_1 -= num_2 # num_1 = num_1 - num_2 = 7 - 2 = 5
print(f'num_1 -= num_2 = {num_1}') # Substraction assignment

num_1: int = 7
num_2: int = 2
num_1 *= num_2  # num_1 = num_1 * num_2 = 7 * 2 = 14
print(f'num_1 *= num_2 = {num_1}') # Multiplication assignment
```

```python
num_1: int = 7
num_2: int = 2
num_1 /= num_2  # num_1 = num_1 / num_2 = 7 / 2 = 3.5
print(f'num_1 /= num_2 = {num_1}') # Division assignment

num_1: int = 7
num_2: int = 2
num_1 %= num_2  # num_1 = num_1 % num_2 = 7 % 2 = 1
print(f'num_1 %= num_2 = {num_1}') # Remainder assignment

num_1: int = 7
num_2: int = 2
num_1 **= num_2  # num_1 = num_1 ** num_2 = 7 ** 2 = 49
print(f'num_1 **= num_2 = {num_1}') # Remainder assignment
```

```python
# 1. Calculating the Area of a Rectangle
# Normal formula: Area = L x B
# Programming: Area = L * B
L: int = 6
B: int = 5
area = L * B
print("Area of the rectangle = ", area)
```

```python
# 2. Calculating the Perimeter of a Rectangle
# Normal formula: Perimeter = 2(L + B)
# Programming: Perimeter = 2 * (L + B)
L: int = 6
B: int = 5
perimeter = 2 * (L + B)
print("Perimeter of the rectangle = ", perimeter)
```

```python
# 5. Calculating the Area of a Circle
# Normal formula: Area = pi x r^2
# Programming: Area = pi * r**2 or Area = pi * r * r
pi: float = 22/7
r: int = 7
area = pi * r**2
print(f"Area of the Circle = {area}")
```

```python
# 5. Calculating the Circumference(Perimeter) of a Circle
# Normal formula: Area = 2 pi r
# Programming: Area = 2 * pi * r
pi:float = 22/7
r:int = 7
perimeter = 2 * pi * r
print("Circumference of the Circle = ", perimeter)
```
